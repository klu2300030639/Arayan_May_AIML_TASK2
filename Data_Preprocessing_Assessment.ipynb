```python
# Part 1 — NumPy Operations

import numpy as np
import pandas as pd
from google.colab import files

uploaded = files.upload()

dataset = pd.read_csv(next(iter(uploaded)))

print("Dataset Loaded Successfully")
print()

dataset['age'] = pd.to_numeric(
    dataset['age'],
    errors='coerce'
)

dataset['math_score'] = pd.to_numeric(
    dataset['math_score'],
    errors='coerce'
)

dataset['science_score'] = pd.to_numeric(
    dataset['science_score'],
    errors='coerce'
)

math_array = np.array(
    dataset['math_score'].dropna()
)

print("Math Score Array:")
print(math_array)

mean_value = np.mean(math_array)
median_value = np.median(math_array)
maximum_value = np.max(math_array)
minimum_value = np.min(math_array)

print("\nMean:",mean_value)
print("Median:",median_value)
print("Maximum:",maximum_value)
print("Minimum:",minimum_value)

normalized_scores = (
    (math_array-minimum_value)
    /(maximum_value-minimum_value)
)

print("\nNormalized Scores:")
print(normalized_scores)


# Part 2 — Pandas Exploration

print("\nPART 2 : PANDAS EXPLORATION\n")

print("First 5 Rows")
print(dataset.head())

print("\nData Types")
print(dataset.dtypes)

print("\nMissing Values")
print(dataset.isnull().sum())

print("\nStudents Attendance Below 70%")

low_attendance = dataset[
    dataset['attendance'] < 70
]

print(low_attendance)


# Part 3 — Data Preprocessing

print("\nPART 3 : DATA PREPROCESSING\n")

print("Missing Values Before Handling")

print(dataset.isnull().sum())

numeric_columns = dataset.select_dtypes(
    include=np.number
).columns

for column in numeric_columns:

    dataset[column] = dataset[column].fillna(
        dataset[column].mean()
    )

categorical_columns = dataset.select_dtypes(
    include='object'
).columns

for column in categorical_columns:

    dataset[column] = dataset[column].fillna(
        dataset[column].mode()[0]
    )

dataset['exam_date'] = pd.to_datetime(
    dataset['exam_date'],
    errors='coerce',
    dayfirst=True
)

score_columns = [
    'math_score',
    'science_score'
]

for column in score_columns:

    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)

    IQR = Q3 - Q1

    lower_limit = Q1 - (1.5*IQR)
    upper_limit = Q3 + (1.5*IQR)

    dataset[column] = np.where(
        dataset[column] > upper_limit,
        upper_limit,
        dataset[column]
    )

    dataset[column] = np.where(
        dataset[column] < lower_limit,
        lower_limit,
        dataset[column]
    )

print("\nOutliers Handled")

duplicates = dataset.duplicated().sum()

print("\nDuplicate Rows:",duplicates)

dataset.drop_duplicates(
    inplace=True
)

print("Duplicates Removed")

print("\nMissing Values After Handling")

print(dataset.isnull().sum())


# Part 4 — Data Analysis

print("\nPART 4 : DATA ANALYSIS\n")

dataset['average_score'] = (
    dataset['math_score']
    +
    dataset['science_score']
)/2

print("Average Score Added")

top_students = dataset.nlargest(
    5,
    'average_score'
)

print("\nTop 5 Students")

print(top_students)

correlation = dataset[
    [
        'attendance',
        'math_score',
        'science_score'
    ]
].corr()

print("\nCorrelation Matrix")

print(correlation)

gender_average = dataset.groupby(
    'gender'
)[
    [
        'math_score',
        'science_score',
        'average_score'
    ]
].mean()

print("\nAverage Marks by Gender")

print(gender_average)

print("\nCompleted Successfully")
```
